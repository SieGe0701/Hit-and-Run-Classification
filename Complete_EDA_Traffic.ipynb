```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("FINAL_CLEANED_TRAFFIC.csv")
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0','LOCATION','CRASH_DATE','DATE_POLICE_NOTIFIED','INJURIES_UNKNOWN'],axis = 1,inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
categorical = ['TRAFFIC_CONTROL_DEVICE','DEVICE_CONDITION','WEATHER_CONDITION','LIGHTING_CONDITION','FIRST_CRASH_TYPE','TRAFFICWAY_TYPE','ALIGNMENT','ROADWAY_SURFACE_COND','ROAD_DEFECT','REPORT_TYPE'
                ,'CRASH_TYPE','HIT_AND_RUN_I','DAMAGE','PRIM_CONTRIBUTORY_CAUSE','SEC_CONTRIBUTORY_CAUSE','STREET_DIRECTION','STREET_NAME','MOST_SEVERE_INJURY']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df = df[categorical].apply(le.fit_transform)
```

```python
df.head()
```

```python
#Separating the target variables and predictors
y = df['HIT_AND_RUN_I']
x = df.drop(['HIT_AND_RUN_I'],axis = 1)
```

```python
ac_scores_wo_smote = []
pc_scores_wo_smote = []
rc_scores_wo_smote = []
f1_scores_wo_smote = []
roc_scores_wo_smote = []
ac_scores_w_smote = []
pc_scores_w_smote = []
rc_scores_w_smote = []
f1_scores_w_smote = []
roc_scores_w_smote = []
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size = 0.75,random_state = 6)
```

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()
model1.fit(xtrain,ytrain)
ypred = model1.predict(xtest)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
from imblearn.metrics import specificity_score
sns.heatmap(confusion_matrix(ytest,ypred),annot= True,fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
ac_scores_wo_smote.append(accuracy_score(ytest,ypred))
pc_scores_wo_smote.append(precision_score(ytest,ypred))
rc_scores_wo_smote.append(specificity_score(ytest,ypred))
f1_scores_wo_smote.append(f1_score(ytest,ypred))
roc_scores_wo_smote.append(roc_auc_score(ytest,ypred))
```

```python
from xgboost import XGBClassifier
model2 = XGBClassifier()
model2.fit(xtrain,ytrain)
ypred = model2.predict(xtest)
sns.heatmap(confusion_matrix(ytest,ypred),annot= True,fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
ac_scores_wo_smote.append(accuracy_score(ytest,ypred))
pc_scores_wo_smote.append(precision_score(ytest,ypred))
rc_scores_wo_smote.append(specificity_score(ytest,ypred))
f1_scores_wo_smote.append(f1_score(ytest,ypred))
roc_scores_wo_smote.append(roc_auc_score(ytest,ypred))


```

```python
from lightgbm import LGBMClassifier
model3 = LGBMClassifier()
model3.fit(xtrain,ytrain)
ypred = model3.predict(xtest)
sns.heatmap(confusion_matrix(ytest,ypred),annot= True,fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
ac_scores_wo_smote.append(accuracy_score(ytest,ypred))
pc_scores_wo_smote.append(precision_score(ytest,ypred))
rc_scores_wo_smote.append(specificity_score(ytest,ypred))
f1_scores_wo_smote.append(f1_score(ytest,ypred))
roc_scores_wo_smote.append(roc_auc_score(ytest,ypred))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import AdaBoostClassifier

models = []
m1 = AdaBoostClassifier()
models.append(('adaboost',m1))

m2 = RandomForestClassifier()
models.append(('RFC',m2))

model4 = VotingClassifier(estimators = models,voting = 'soft')
model4.fit(xtrain,ytrain)
ypred = model4.predict(xtest)
```
